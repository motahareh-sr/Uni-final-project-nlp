```python
# import stopwords
import re
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
from hazm import stopwords_list

# خواندن کلمات فارسی از فایل persion.txt
# with open('/content/drive/My Drive/project/finals/persian.txt', 'r', encoding='utf-8') as file:
#     persian_stopwords = file.read().split()

english_stopwords = set(stopwords.words('english'))
```

```python
from hazm import word_tokenize
from hazm import Normalizer
from hazm import Lemmatizer

# Function to tokenize and lemmatize sentences
def process_sentences(records):
    dic = []
    normalizer = Normalizer()
    lemmatizer = Lemmatizer()

    for sentence in records:
        normalized_sentence = normalizer.normalize(sentence)
        tokens = word_tokenize(normalized_sentence)
        processed_tokens = []

        for word in tokens:
            if re.search(r'[\u0600-\u06FF]', word):  # Persian
                word = lemmatizer.lemmatize(word)
                if word not in stopwords_list():
                    processed_tokens.append(word)
            elif word.lower() not in english_stopwords:  # English
                processed_tokens.append(word)

        dic.append(processed_tokens)

    return dic

# Process sentences
dic1 = process_sentences(records)
dic2 = process_sentences(records2)
```

```python
from nltk.util import ngrams
from collections import Counter

# Function to compute n-gram counts for each sentence
def get_ngram_counts(dic, max_n):
    ngram_counts = []
    for sentence in dic:
        sentence_ngrams = Counter()
        for n in range(1, max_n + 1):
            grams = list(ngrams(sentence, n))
            sentence_ngrams.update(Counter(grams))
        ngram_counts.append(sentence_ngrams)
    return ngram_counts

# Compute n-gram counts
max_n = 5
ngram_counts1 = get_ngram_counts(dic1, max_n)
ngram_counts2 = get_ngram_counts(dic2, max_n)
```

```python
# Function to compute similarity between two sentences
def compute_similarity(counts1, counts2):
    common_ngrams = set(counts1.keys()) & set(counts2.keys())
    common_count = sum(min(counts1[gram], counts2[gram]) for gram in common_ngrams)
    total_count = sum(counts1.values()) + sum(counts2.values())
    similarity = (2 * common_count) / total_count if total_count != 0 else 0
    return similarity
```

```python
# find most similar one
similarity_results = []

for idx1, counts1 in enumerate(ngram_counts1):
    max_similarity = 0
    most_similar_idx = -1
    for idx2, counts2 in enumerate(ngram_counts2):
        similarity = compute_similarity(counts1, counts2)
        if similarity > max_similarity:
            max_similarity = similarity
            most_similar_idx = idx2
    similarity_results.append((records[idx1], records2[most_similar_idx], max_similarity))

df_results = pd.DataFrame(similarity_results, columns=['Record from List 1', 'Most Similar Record from List 2', 'Similarity'])

# Display the results
df_results.head()
```
